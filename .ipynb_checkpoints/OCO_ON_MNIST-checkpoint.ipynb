## Unconstrained gradient descent

```python
def hingereg(x,a,b, Lambda):
    """ hinge loss function """
    threshold = (a @ x) * b
    cost = 1- (a  @ x) * b
    cost[threshold >= 1] = 0
    return cost.mean() + 0.5* Lambda * (x**2).sum()

def  gradreg(x, a, b,Lambda):
    """gradient of the hinge loss"""
    threshold =  b * (a @ x) # define hard-margin SVM
    gradient = - b[:, np.newaxis] * a
    gradient[threshold >= 1] = 0
    return(gradient.mean(axis = 0)+Lambda*x) # standardized dividing by n
  
```

```python
import time
def GD(a, b, init,iters, cost,  grad, Lambda=1, verbose = False):
    """
    Unconstrained gradient descent function
    """
    params = [init]
    costs = []
    times= []
    for i in range(iters):
        s = time.time()
        eta = 1 /(Lambda * (i+1))
        param = params[-1] - eta * grad(params[-1], a, b, Lambda)
        params.append(param)
        c = cost(params [-1], a ,b ,Lambda)
        costs.append(c)
        times.append(time.time() - s)
    if verbose:    
        print("Final cost: "+ str(costs[-1]) +"\n")
        print("Parameters:" + str(params[-1]) +" ")

    params  = np.array(params)
    costs = np.array(costs)
    return params , costs, times
```

### function returning error rate on test set

```python
def test(params):
    scores = []
    for i in params:
        scores.append( ((2*( X_test@i >= 0)-1) != Y_test).mean() )
    return scores
```

## L1 projection

```python
import pdb #for debuging


def projection_l1(x, z=1):
    u = np.abs(x) 
    if u.sum() > z:
        proj = projection_simplex(u, z)
        proj *= np.sign(x)
        return proj
    return x

def projection_simplex(v, z=1):
    n, = v.shape  
    if v.sum() == z and np.alltrue(v >= 0):
        return v
    u = np.sort(v)[::-1]
    cssv = np.cumsum(u)
    rho = np.nonzero(u * np.arange(1, n+1) > (cssv - z))[0][-1]
    theta = float(cssv[rho] - z) / rho
    w = (v - theta).clip(min=0)
    return w



```

# GD projected on the $\mathcal{l}_1$ ball

```python
# GD projected on the l1 ball

def GDproj(a, b, init, iters, cost,  grad, Lambda=1, z=1, verbose = False):
    params = [init]
    costs = []
    times = []
    for i in range(iters):
        s = time.time()
        eta = 1 /(Lambda * (i+1))
        param = projection_l1(params[-1] - eta * grad(params[-1], a, b, Lambda), z)# projection of the updated params
        params.append(param)
        c = cost(params [-1], a ,b ,Lambda)
        costs.append(c)
        times.append(time.time() - s)
    if verbose:    
        print("Final cost: "+ str(costs[-1]) +"\n")
        print("Parameters:" + str(params[-1]) +" ")

    params  = np.array(params)
    costs = np.array(costs)
    return params , costs, times
```

# SGD

```python
from random import randint
import random

def sample(a, b, T):
    """
        Sample a list of T integers uniformly from {a,..,b}
    """
    return [randint(a, b) for i in range(T)]
```

```python
def sample_grad(x, a, b, Lambda, i):
    """
        gradient wrt the ith data point
    """
    grad = Lambda * x
    if (a[i] @ x.T)*b[i] < 1 :
        grad = grad-b[i]*a[i]
    return grad
```

## SGD proj implementation

```python
def SGD(a, b, init, iters, cost,  grad, Lambda=1, z=1, verbose = False):
    """
    Stochastic gradient descent function
    """
    # initialization
    params = [init]
    costs = []
    c = cost(params[-1], a ,b ,Lambda)
    costs.append(c)
    
    x = init

    # sample uniformly the gradients that will be used in all iterations
    uniform_samples = sample(0, a.shape[0]-1, iters)
    
    times = []
    for i in range(iters):
        s = time.time()
        
        # algorithm steps
        eta = 1 /(Lambda*(i+1)) # leraning rate 
        x = x - eta * grad(x, a, b, Lambda, uniform_samples[i]) #gradient step
        x = projection_l1(x, z) # projection on the l1-ball
        times.append(time.time() - s)
        """
        # logging
        param = ((i)*params[-1] + x)/(i+1) # averaging
        params.append(param)        
        c = cost(params[-1], a ,b ,Lambda)
        costs.append(c)
        """
        
        # log the current param and instant cost 100 times across ietrs
        if ((i+1) % (iters/100)) == 0 :
            param = ((i)*params[-1] + x)/(i+1) # averaging
            params.append(param)        
            c = cost(params[-1], a ,b ,Lambda)
            costs.append(c)
    if verbose:    
        print("Final cost: "+ str(costs[-1]) +"\n")
        print("Parameters:" + str(params[-1]) +" ")

    params  = np.array(params)
    costs = np.array(costs)
    return params , costs, times
```

# Stochastic exponentiated Gradient +/-

```python
def SEG_plus_moins(a, b, init, iters, cost,  grad, Lambda=1, z=1, verbose = False):
    """
    Stochastic Mirror descent function
    """
    # initialization
    params = [np.zeros(a.shape[1])]
    costs = []
    c = cost(params[-1], a ,b ,Lambda)
    costs.append(c)
    
    x = np.zeros(a.shape[1])
    w = np.ones(a.shape[1]) / (2 * a.shape[1])
    theta = np.zeros(2 * a.shape[1])

    # sample uniformly the gradients that will be used in all iterations
    uniform_samples = sample(0, a.shape[0]-1, iters)
    times = []
    for i in range(iters):
        s = time.time()
        # algorithm steps
        eta = 1 /np.sqrt((i+1)) # learning rate 
        theta[:a.shape[1]] = theta[:a.shape[1]]  - eta * grad(x, a, b, Lambda, uniform_samples[i])
        theta[a.shape[1]:] = theta[a.shape[1]:]  + eta * grad(x, a, b, Lambda, uniform_samples[i])
        
        w = np.exp(theta) / np.exp(theta).sum()
        
        x = z * ( w[:a.shape[1]] - w[a.shape[1]:])
        times.append(time.time()-s)
        #x = projection_l1(yt, z) # projection on the l1-ball
        
        # log the current param and instant cost 100 times across ietrs
        if ((i+1) % (iters/100)) == 0 :
            param = ((i)*params[-1] + x)/(i+1) # averaging
            params.append(param)        
            c = cost(params[-1], a ,b ,Lambda)
            costs.append(c)
    if verbose:    
        print("Final cost: "+ str(costs[-1]) +"\n")
        print("Parameters:" + str(params[-1]) +" ")

    params  = np.array(params)
    costs = np.array(costs)
    return params , costs, times
```

# ADAgrad
## weighted projection on the l1 ball

```python
def l1_proj_weight(x,z=1,d=0):
    if sum(abs(x)) <= z:
        return x
    else:
        w = proj_weight(abs(x)/z,d)
        return z*np.sign(x)*w

def proj_weight(x,D):
    dx = np.abs(np.dot(D,x))
    sorted_indices = np.argsort(-dx, kind='quicksort')
    sx = np.cumsum(x[sorted_indices])
    sd = np.cumsum(1/np.diag(D)[sorted_indices])
    res = dx[sorted_indices] - (sx-1)/sd
    d0 = np.max(np.where(res>0)[0])
    theta = (sx[d0]-1)/sd[d0]
    return np.maximum(0,dx-theta)
```

## AdaGrad Implementation

```python
def ADAGRAD(a, b, init, iters, cost,  grad, Lambda=1, z=1, verbose = False):
    """
    ADAGRAD
    """
    # initialization
    params = [np.zeros(a.shape[1])]
    costs = []
    c = cost(params[-1], a ,b ,Lambda)
    costs.append(c)
    x = y = np.zeros(a.shape[1])
    S = np.zeros(a.shape[1]) + 0.01
    

    # sample uniformly the gradients that will be used in all iterations
    uniform_samples = sample(0, a.shape[0]-1, iters)
    times = []
    for i in range(iters):
        start = time.time()
        gradient = grad(x, a, b, Lambda, uniform_samples[i])
        S = S + np.square(gradient)
        D = np.diag(np.sqrt(S))
        Dinvers = np.diag(1/np.sqrt(S))
        y = x - Dinvers @ gradient
        x = l1_proj_weight(y, z, D)
        times.append(time.time() -start)
        
        # log the current param and instant cost 100 times across ietrs
        if ((i+1) % (iters/100)) == 0 :
            param = ((i)*params[-1] + x)/(i+1) # averaging
            params.append(param)        
            c = cost(params[-1], a ,b ,Lambda)
            costs.append(c)
    if verbose:    
        print("Final cost: "+ str(costs[-1]) +"\n")
        print("Parameters:" + str(params[-1]) +" ")
    params  = np.array(params)
    costs = np.array(costs)
    return params , costs, times
```

# Stochastic Randomized Exponentiated Gradient $\pm$ :

```python
def SREG(a, b, init, iters, cost,  grad, Lambda=1, z=1, verbose = False):
    """
    Stochastic Randomized Exponentiated Gradient +_
    """

    # dimention of the problem
    d = a.shape[1]
    # MAB fashion in a similar way to SEG
    w = np.repeat(1/(2*d), 2*d)    
    # initialization
    start_time = time.time()
    times = [0]
    x = z*(w[0:d]-w[d:2*d])
    param = x
    params = [param]
    costs = []
    c = cost(params[-1], a ,b ,Lambda)
    costs.append(c)
    
    # sample uniformly the gradients that will be used in all iterations
    random.seed(0)
    uniform_samples = sample(0, a.shape[0]-1, iters)
    
    for i in range(iters):
        # algorithm steps
        eta = np.sqrt(np.log(2*d)/(2*d*(i+1))) # leraning rate 
        sampled_gradient = grad(x, a, b, Lambda, uniform_samples[i]) #compute the gradient on the selected datapoint
        
        # sample a coordinate/direction (an expert)
        coord = np.random.multinomial(1, ((w[0:d]+w[d:2*d])/2), size=1).argmax()
        sampled_gradient_coord = sampled_gradient[coord] # we only observe the gradient on this direction
        
        # gradient step
        if sampled_gradient_coord>0 :
            if w[coord] != 0:
                w[coord] = np.exp((-eta*sampled_gradient_coord)/w[coord])*w[coord]
            else:
                w[coord] = 0
        else:
            if w[coord + d] != 0:
                w[coord+d] = np.exp((eta*sampled_gradient_coord)/w[coord+d])*w[coord+d]
            else:
                w[coord] = 0
        w = w / (w.sum())
        x = z*(w[0:d]-w[d:2*d])
        
        """
        # logging
        param = ((i)*params[-1] + x)/(i+1) # averaging
        params.append(param)        
        c = cost(params[-1], a ,b ,Lambda)
        costs.append(c)
        """
        
        # log the current param and instant cost 100 times across ietrs
        if ((i+1) % (iters/100)) == 0 :
            times.append(time.time()-start_time)
            param = ((i)*params[-1] + x)/(i+1) # averaging
            params.append(param)        
            c = cost(params[-1], a ,b ,Lambda)
            costs.append(c)
            
        if verbose:    
            print("Final cost: "+ str(costs[-1]) +"\n")
            print("Parameters:" + str(params[-1]) +" ")

    params  = np.array(params)
    costs = np.array(costs)
    return params , costs, times
```

```python
import copy
def ONS(a, b, init, iters, cost,  grad, Lambda=1, z=1,gamma=1, verbose = False):
    """
    ONS
    """
    # initialization
    params = [np.zeros(a.shape[1])]
    costs = []
    A = (gamma**2)*np.identity(a.shape[1])
    Ainv = np.linalg.inv(A)
    
    c = cost(params[-1], a ,b ,Lambda)
    costs.append(c)
    x = np.zeros(a.shape[1])
    y = copy.deepcopy(x)
    # sample uniformly the gradients that will be used in all iterations
    uniform_samples = sample(0, a.shape[0]-1, iters)
    times = []
    for i in range(iters):
        start = time.time()
        gradient = grad(x, a, b, Lambda, uniform_samples[i])
        A = A + gradient@gradient.T + Lambda * x
        coef = (1+ gradient.T@Ainv@gradient)
        Ainv = Ainv - ((Ainv@gradient) @ (gradient.T@Ainv)) /coef
        y = x - (1/gamma) * Ainv@gradient
        x = l1_proj_weight(y, z, d=np.diag(np.diag(A)))
        times.append(time.time() -start)
        
        # log the current param and instant cost 100 times across ietrs
        if ((i+1) % (iters/100)) == 0 :
            param = ((i)*params[-1] + x)/(i+1) # averaging
            params.append(param)        
            c = cost(params[-1], a ,b ,Lambda)
            costs.append(c)
    if verbose:    
        print("Final cost: "+ str(costs[-1]) +"\n")
        print("Parameters:" + str(params[-1]) +" ")
    params  = np.array(params)
    costs = np.array(costs)
    return params , costs, times
```

```python
import pdb
# projection for a vector v with non-negative coordinates on the simplex of radius z

def projection(x, z=1):
    u = np.abs(x) 
    if u.sum() > z:
        proj = l1_projection(u, z)
        proj *= np.sign(x)
        return proj
    return x

def l1_projection(v, z=1):
    n, = v.shape  
    if v.sum() == z and np.alltrue(v >= 0):
        return v
    u = np.sort(v)[::-1]
    cssv = np.cumsum(u)
    rho = np.nonzero(u * np.arange(1, n+1) > (cssv - z))[0][-1]
    theta = float(cssv[rho] - z) / rho
    w = (v - theta).clip(min=0)
    return w
```

```python
def multiple_sample(a, b, T, m):
    """
        Sample a list of T integers uniformly from {a,..,b}
    """
    return np.random.choice(np.arange(a,b), size=(T, m), replace=True)
```

```python
def SVRG(a, b, init, iters, cost,  grad_full, grad_sample, option = 2, m = 10 ,Lambda=1, verbose = False):
    """
    Stochastic gradient descent function
    """
    eta = 1/iters
    # initialization of logging variables
    start_time = time.time()
    times = [0]
    params = []
    costs = []
    
    ws_tilde = [init]
    """
    costs = []
    c = cost(params[-1], a ,b ,Lambda)
    costs.append(c)
    """
    w_tilde = ws_tilde[-1]

    # sample uniformly the gradients that will be used in all iterations
    uniform_samples = multiple_sample(0, a.shape[0], iters, m)
    # sample uniformly the updates of w_tilde from w for all iterations
    w_tilde_samples = sample(0, m-2, iters)
    
    for s in range(iters):
        # w_tilde
        w_tilde = ws_tilde[-1]
        mu_tilde = grad_full(w_tilde.T, a, b, Lambda) # gradient on w_tilde
        # w
        ws = [w_tilde]
        w = ws[-1]
        
        # logging the cost and time
        params.append(w)
        c = cost(params[-1], a ,b ,Lambda)
        costs.append(c)
        times.append(time.time()-start_time)
        
        for t in range(m):
            i_t = uniform_samples[s][t] # sample uniformly the gradient
            grad_i_t_w_tilde = grad_sample(w_tilde, a, b, Lambda, i_t)
            grad_i_t_w = grad_sample(w, a, b, Lambda, i_t)
            w = w - eta*(grad_i_t_w - grad_i_t_w_tilde + mu_tilde)
            ws.append(w)
        if option == 1 :
            ws_tilde.append(ws[-1])
        elif option == 2 :
            ws_tilde.append(ws[w_tilde_samples[s]])
        else:
            raise ValueError('error no such option exists')
    if verbose:    
        print("Final cost: "+ str(costs[-1]) +"\n")
        print("Parameters:" + str(params[-1]) +" ")

    params  = np.array(params)
    costs = np.array(costs)
    return params , costs, times
```

```python
# for the sake of verifying empirical results of SVRG let's define and SGD wita fixed learning rate
def SGD_testing(a, b, init, iters, cost,  grad, eta_mode, Lambda=1, z=1, verbose = False):
    """
    Stochastic gradient descent function
    """
    # initialization
    start_time = time.time()
    times = [0]
    params = [init]
    costs = []
    c = cost(params[-1], a ,b ,Lambda)
    costs.append(c)
    
    x = init
    
    
    if eta_mode == 'fixed':
        eta = 1 /iters # leraning rate 

    # sample uniformly the gradients that will be used in all iterations
    uniform_samples = sample(0, a.shape[0]-1, iters)
    
    for i in range(iters):
         
        if eta_mode == 'decay' :
            eta = 1 /(Lambda*(i+1)) # leraning rate
        # algorithm steps
        x = x - eta * grad(x, a, b, Lambda, uniform_samples[i]) #gradient step
        x = projection(x, z) # projection on the l1-ball     

        times.append(time.time() - start_time)
        #param = ((i)*params[-1] + x)/(i+1) # averaging
        param = x
        params.append(param)        
        c = cost(params[-1], a ,b ,Lambda)
        costs.append(c)
    if verbose:    
        print("Final cost: "+ str(costs[-1]) +"\n")
        print("Parameters:" + str(params[-1]) +" ")

    params  = np.array(params)
    costs = np.array(costs)
    return params , costs, times
```
